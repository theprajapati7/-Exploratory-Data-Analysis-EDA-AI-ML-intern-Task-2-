### HISTOGRAMS

```python
df.hist(figsize=(12,10),bins=20)
plt.suptitle("Histogram of Numeric Features")
plt.tight_layout()
plt.show()        
```

### BOXPLOT

```python
numeric_cols = df.select_dtypes(include='number').columns

plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols) // 3 + 1, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

### PAIRPLOT

```python
sns.pairplot(df[["Age","Fare","SibSp","Parch"]], corner=True)
plt.suptitle("Pairplot",y=1.02)
plt.show()
             
```

### TRENDS IN NUMERIC FEATURES

```python
df[['Fare']].plot(kind='line', figsize=(10, 6))
plt.title("Trend in Fare")
plt.ylabel("Fare")
plt.show()
```

### SCATTER PLOT FOR FEATURE RELATIONSHIPS

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Scatterplot: Age vs Fare")
plt.show()
```

### BOXPLOT TO DETECT OUTLIAR

```python
sns.boxplot(x="Fare",data=df)
plt.title("Boxplot of fare")
plt.show()
```

### HISTOGRAM TO CHECK DISTRIBUTION

```python
df["Fare"].hist(bins=20,figsize=(10,6))
plt.title("Gistogram of fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

### BARPLOT FOR PATTERNS

```python
sns.barplot(x='Pclass',y='Fare',data=df)
plt.title('Avg. fare by Pclass')
plt.show()
```

### FEATURE LEVEL INFERENCE

Based on the visualizations and summary statistics, I can draw the following insights about the dataset:

### Fare
- The distribution of `Fare` is right-skewed, meaning most passengers paid low fares while a few paid extremely high fares.
- The median fare is much lower than the maximum, suggesting a concentration of passengers in lower fare ranges.
- Outliers are present, especially at the higher end (fare > 100), which might represent first-class luxury travelers.

### Pclass vs Fare
- Passengers in Pclass 1 paid significantly higher fares compared to those in Pclass 2 and 3.
- This indicates a strong correlation between passenger class and ticket cost.

### Age
- The `Age` distribution is fairly normal but slightly right-skewed with a majority of passengers aged between 20 and 40.
- Some outliers are present above age 70.
- The dataset contains some missing or zero values in `Age`, which may need to be handled during preprocessing.

### Age vs Fare (Scatterplot)
- There is a weak positive relationship between `Age` and `Fare`â€”older passengers are slightly more likely to have paid higher fares.
- However, there is no strong linear trend, suggesting other factors influence ticket price more.

### Outliers
- Several numeric features like `Fare` and `Age` show clear outliers, especially in boxplots.
- These outliers might impact model performance and can be either treated or kept depending on the analysis goal.

### Overall Trends
- Most passengers belonged to lower classes and paid lower fares.
- A few high-paying passengers may be influential points in downstream modeling or segmentation.
